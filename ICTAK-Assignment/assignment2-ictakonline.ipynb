```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Create a NumPy array of shape (5, 5) with random integers between 1 and 50.
#       Find the row with the maximum sum.
#       Replace all elements greater than the mean with 0.
```

```python
# arr = np.arange.rand((25).reshape(5, 5)))
arr = np.random.randint(1, 51, size=(5, 5))

print("Original 5x5 Array:")
print(arr)
```

```python
row_sums = arr.sum(axis=1)
max_sum_row_index = np.argmax(row_sums)
print(f"Row sums: {row_sums}")
print(f"Row with the maximum sum (Index {max_sum_row_index}): {arr[max_sum_row_index]}")
```

```python
array_mean = arr.mean()
print(f"Mean of the array: {array_mean:.2f}")
```

```python
arr_modified = np.where(arr > array_mean, 0, arr)
print("\nArray after replacing elements grater mean with 0:")
print(arr_modified)
```

```python
# 3. Load the dataset Iris data
#       Display the first 10 rows and basic statistics.
#       Filter all rows where sepal_length > median and species is "versicolor".
#       Group the data by species and compute the mean petal_length.
```

```python
iris_df = sns.load_dataset('iris')

print("First 10 rows of the Iris dataset:")
print(iris_df.head(10))
```

```python
print("basic Statistics:")
print(iris_df.describe())
```

```python
median_sepal_length = iris_df['sepal_length'].median()
print(f"Median of sepal_length: {median_sepal_length}")
```

```python
filtered_df = iris_df[(iris_df['sepal_length'] > median_sepal_length) & (iris_df['species'] == 'versicolor')]
print(f"Filtered rows (sepal_length > median and species == 'versicolor'): {filtered_df}")
```

```python
mean_petal_length = iris_df.groupby('species')['petal_length'].mean()
print(f"Mean petal_length per species: {mean_petal_length}")
```

```python
# 4. Add a new column named petal_ratio = petal_length / petal_width.
#       Find the top 5 rows with the highest petal_ratio.
```

```python
iris_df['petal_ratio'] = iris_df['petal_length'] / iris_df['petal_width']
print(f"Iris dataset with new 'petal_ratio' column (showing last 5 rows): {iris_df.tail()}")
# print(iris_df.tail())
```

```python
top_5_ratio = iris_df.sort_values(by='petal_ratio', ascending=False).head(5)
print (f"top 5 rows with the highest petal ratio: {top_5_ratio}")
```

```python
# 5. Using Matplotlib, create:
#       A histogram of sepal_length with 15 bins.
#       A scatter plot of sepal_length vs petal_length, color-coded by species.
```

```python
plt.figure(figsize=(8, 5))
plt.hist(iris_df['sepal_length'], bins=15, color='blue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

sns.scatterplot(data=iris_df, x='sepal_length', y='petal_length', hue='species', palette='bright')
plt.title('sepal Length vs. peatal Length')
plt.xlabel('sepal Length (cm)')
plt.ylabel('Paetal Length (cm)')
plt.legend(title='Species')
plt.grid(True)
plt.show()
```

```python
# 6. Using Seaborn, plot:
#       A pairplot of the entire dataset.
#       A boxplot showing petal_width distribution for each species.
```

```python
sns.pairplot(iris_df, hue='species', palette='viridis')
plt.suptitle('Pairplot of Iris Dataset') 
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_width', data=iris_df)

plt.title('petal Width Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Peaal Width (cm)')
plt.grid(axis='y', linestyle='--')
plt.show()
```
