## Use Heart Disease [Dataset](https://github.com/cksajil/DSAIRP25/blob/main/datasets/heart_disease.csv) and answer the following questions

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('heart_disease.csv')
```

```python
df
```

1. Find the top 5 important features to the target column

```python
corr_matrix = df.corr()
corr_matrix
```

```python
target_corr = corr_matrix['target'].abs().sort_values(ascending=False)
target_corr
```

```python
top_5_features = target_corr[1:6]
print("top 5 most important features related to target is:")
print(top_5_features)
```

2. Perform Box-Cox Transformations to relevant features

```python
from scipy.stats import boxcox
```

```python
df
```

```python
numerical_cols = ['age', 'trestbps', 'chol', 'thalach']
```

```python
for col in numerical_cols:
    if (df[col] > 0).all():
        df[col + '_boxcox'], _ = boxcox(df[col])
```

```python
df
```

3. Perform Feature Binning to Age Column and add it as a new column to the dataset

```python
df
```

```python
age_bins = [0, 20, 30, 55, 70, 100]
age_labels = ['Teens','Youth', 'Middle-Aged', 'Senior', 'VeryOld']
```

```python
df['AgeGroup'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)
display(df)
```

```python
print("Feature binning applied to 'age' column. New column is 'AgeGroup'.")
print("Value counts for the new 'AgeGroup' column:")
print(df['AgeGroup'].value_counts())
```
